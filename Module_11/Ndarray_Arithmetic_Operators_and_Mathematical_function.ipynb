## ndarray Arithmetic Operators and Mathematical function

```python
import numpy as np 

x = np.array([10,20,30,40,50])
y = np.array([1,2,3,4,5])

z = x + y
print(z)

z = x - y
print(z)

z = x * y
print(z)

z = x / y
print(z)

z = x % y
print(z)

z = x ** y
print(z)

z = x // y
print(z)
```

```python
print("sin")
sin = np.sin(x)
print(sin)

print("cos")
cos = np.cos(x)
print(cos)

print("tan")
tan = np.tan(x)
print(tan)

print("sqrt")
sqrt = np.sqrt(x)
print(sqrt)

print("log")
log = np.log(x)
print(log)

print("deg_conv")
deg_conv = np.deg2rad(x)
print(deg_conv)

print("rad_conv")
rad_conv = np.rad2deg(x)
print(rad_conv)

print("exp")
exp = np.exp(x)
print(exp)

print("log10")
log10 = np.log10(x)
print(log10)

print("sum")
sum = np.sum(x)
print(sum)

print("mean")
mean = np.mean(x)
print(mean)

print("median")
median = np.median(x)
print(median)

print("std")
std = np.std(x)
print(std)

print("var")
var = np.var(x)
print(var)

print("max")
max = np.max(x)
print(max)

print("min")
min = np.min(x)
print(min)

print("range")
range = np.max(x) - np.min(x)
print(range)
```
