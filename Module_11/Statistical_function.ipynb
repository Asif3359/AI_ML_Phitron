## Statistical function

```python
import numpy as np

df = np.genfromtxt('student_dataset.csv',delimiter=',',skip_header=1)
# remove the first and second column
df = df[:,2:]
print(df)
```

```python
# statistical function

from numpy import median


math_marks = df[:,0]
science_marks = df[:,1]
english_marks = df[:,2]
grade_level = df[:,3]
study_hour = df[:,4]

print(math_marks)
print(science_marks)
print(english_marks)
print(grade_level)
print(study_hour)

max_math_marks = np.max(math_marks)
print(max_math_marks)

min_math_marks = np.min(math_marks)
print(min_math_marks)

mean_math_marks = np.mean(math_marks)
print(mean_math_marks)

median_math_marks = np.median(math_marks)
print(median_math_marks)

std_math_marks = np.std(math_marks)
print(std_math_marks)


corr_math_marks = np.corrcoef(grade_level,study_hour)
print(corr_math_marks)
```
