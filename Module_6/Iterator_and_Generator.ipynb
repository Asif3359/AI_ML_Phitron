```python
# Iterator

set = {10,30,20,30,80,40,45}

it = iter(set)

# print(next(it))

for i in it :
    print(i, end=" ")

```

```python
# Generator

lst = [x for x in range(501)]

def data_loader(chunk_size, lst) :
    for i in range (0,len(lst),chunk_size) :
        yield lst[i:i+chunk_size]

x = data_loader(5,lst)

print(next(x))

print()
for i in x :
    print(i)

        
# print(lst)
```
