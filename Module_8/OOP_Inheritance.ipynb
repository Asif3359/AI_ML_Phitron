```python
# OOP Inheritance
# instance attributes 
class Phone :
    # constructure 
    def __init__(self,model,battery,camera, battery_percentage=100):
        self.model = model
        self.battery = battery
        self.camera = camera
        self.battery_percentage = battery_percentage
    # methods 
    def charge(self , hour):
        if self.battery_percentage >=100 :
            print(f"charhge completed")
        else :
            self.battery_percentage += hour 
            print(f"charhge completed by {hour} hr")
    def capture(self, total_photo):
        if (self.battery_percentage <= 0 ) :
            print("No Photo")
        else:
            self.battery_percentage -= total_photo
            print(f"Photo captured in {self.model}")

class Cooling :
    # constructor
    def __init__(self, cooling_method):
        self.cooling_method = cooling_method
    # methods
    def cooling_on (self):
        print(f"The system is cool by {self.cooling_method}")

apple = Phone(model="Apple 14",battery="1000mh", camera = "40Mpx")

apple.model = "Apple 15"

print(apple.model)
print(apple.battery)
print(apple.camera)
print(apple.battery_percentage)

apple.capture(2)
print(apple.battery_percentage)
apple.charge(2)
print(apple.battery_percentage)




print()
motorola = Phone2(model="motorola 14",battery="3000mh", camera = "30Mpx")

motorola.model = "motorola 15"

print(motorola.model)
print(motorola.battery)
print(motorola.camera)
print(motorola.battery_percentage)
print(motorola.battery_percentage)
```

```python
class SmartPhone (Phone) :
    def __init__(self,model,battery,camera, processor):
        super().__init__(model,battery,camera)
        self.processor = processor
    def charge (self,hour):
        if self.battery_percentage >=100 :
            print(f"fast charhge completed")
        else :
            self.battery_percentage += hour+1 
            print(f"fast charhge completed by {hour} hr")
        super().charge(hour)


pro = SmartPhone(model="mt 14",battery="3000mh", camera = "30Mpx", processor="SnapDragon")

print(pro.model)
print(pro.battery)
print(pro.camera)
print(pro.battery_percentage)
print(pro.processor)


pro.capture(2)
print(pro.battery_percentage)
pro.charge(1)
print(pro.battery_percentage)
```

```python
# Multiple inheritance 

class Smart_with_cool(SmartPhone, Cooling):
    def __init__(self,model,battery,camera, processor, cooling_method):
        SmartPhone.__init__(self,model,battery,camera, processor)
        Cooling.__init__(self,cooling_method)

pro_cooling = Smart_with_cool(
    model="mt 14",
    battery="3000mh", 
    camera = "30Mpx", 
    processor="SnapDragon",
    cooling_method="H20"
)

print(pro_cooling.model)
print(pro_cooling.battery)
print(pro_cooling.camera)
print(pro_cooling.battery_percentage)
print(pro_cooling.processor)
print(pro_cooling.cooling_method)
pro_cooling.cooling_on()


pro_cooling.capture(2)
print(pro_cooling.battery_percentage)
pro.charge(1)
print(pro_cooling.battery_percentage)
```
